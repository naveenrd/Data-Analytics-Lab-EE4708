** How many bank names start with the letter 's' ? **

```python
# CODE HERE
```

```python
count=0
for i in range(len(banks)):
    if banks.iloc[i]['Bank Name'][0]=='S':
        count+=1
```

** How many bank names consist of just two words? (e.g. "First Bank" , "Bank Georgia" )**

```python
# CODE HERE
```

```python
count=0
for i in range(len(banks)):
    if len(banks.iloc[i]['Bank Name'].split())==2:
        count+=1
```

```python
print('number of banks consist of just 2 words: ',count)
```

```python
import datetime
```

```python
def convert_to_datetime(string):
    return datetime.datetime.strptime(string,'%d-%b-%y').year
```
